 ## 2.2 Missing values Check

```python
# Define null summary function and % of null value

def display_null_summary(df):
    """
    Display the count and percentage of null values for each column in a DataFrame.

    Parameters:
        df (pd.DataFrame): The input DataFrame to check for null values.

    Returns:
        None: Displays a table of null counts and percentages.
    """
    # Calculate null counts and percentages
    null_counts = df.isnull().sum()
    pct_null = (null_counts / len(df)) * 100
    
    # Combine into a DataFrame for display
    null_summary = pd.DataFrame({
        'Null Records': null_counts,
        'Percentage Null': pct_null.round(2)
    })
    
    # # Filter only columns with null values
    # null_summary = null_summary[null_summary['Null Records'] > 0]
    
    # Display the result
    if null_summary.empty:
        print("No null values found in the DataFrame.")
    else:
        print("Summary of Null Records and Percentages:")
        display(null_summary)
```

```python
# plot distribution and boxplot by new function definition
def plot_distribution_and_outliers(df, column):
    # Create a figure with 1 row and 2 columns - use subplot to create mulitple graph in a layout  
    fig, axes = plt.subplots(1,2, figsize = (12,5))  # fig: figure object, axes: array of each axes objects for each graph with a table of 1 rows and 2 columns
    
    # Histogram for Distribution on axes[0] (draw graph on left axes)
    sns.histplot(df[column], bins=20, kde = True, color = 'skyblue', ax=axes[0])
    axes[0].set_title(f'{column}-Distribution')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('frequency')
    axes[0].grid(True) # to visualize the gridline

    # Boxplot for outlier on right graph on axes[1] (draw graph on right axes)
    sns.boxplot(x=df[column], color = 'lightcoral', ax=axes[1])
    axes[1].set_title(f'{column}-Outliers')
    axes[1].set_xlabel(column)
    axes[1].grid(True)
    
    
    # Adjust layout for better appearance
    plt.tight_layout()

plot_distribution_and_outliers(train_data, 'Age')
```

```python
# Calculate survival rate by Pclass
pclass_survival = train_data.groupby('Pclass')['Survived'].mean()
print("Survival Rate by Pclass")
print(pclass_survival)
```

```python
# Barplot for Survival Rate by Pclass
plt.figure(figsize = (8,6))
sns.barplot(x='Pclass', y='Survived',data=train_data, ci = None)
plt.title('Survival Rate by Pclass')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Survival Rate')
```

The Upper SES tends to have more survival rate than the lower class.

This disparity can be attributed to several factors:
- Proximity to Lifeboats: First-class passengers had cabins closer to the lifeboats, allowing them quicker access during the evacuation.
- Crew Assistance: Passengers from higher classes often received more attention and assistance from the crew.
- Physical Barriers: Third-class passengers faced challenges such as locked gates and complex passageways, delaying their escape to safety.

```python
# Create contingency table
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab( train_data['Pclass'], train_data['Survived']) # rows x column

# Perform Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Results
print('Chi-Square Test Results:')
print(f'Chi-Square Statistic: {chi2_stat:.4f}')
print(f'p-value:{p_value:.4f}')
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
if p_value < 0.05:
    print("There is a significant relationship between Pclass and Survival")
else: 
    print("There is no significant relationship between Pclass and Survival")
```

**Chi-Square results:**

Observed Frequencies: The actual survival counts in each passenger class (Pclass).
Expected Frequencies: The counts we would expect if there were no relationship between class and survival.

Since the Chi-Square value is large, it suggests that the survival rates vary significantly between passenger classes. This means Pclass influences survival—for example, passengers in higher classes (like first class) may have had a better chance of survival compared to lower classes.

The survival distribution is not random with respect to Pclass; instead, **it's shaped by the class of the passengers.**

**P-value results:**

p-value: A p-value of 0.0000 (less than 0.05) indicates that the relationship between Pclass and Survival is statistically significant. This means we reject the null hypothesis that there is no relationship between the two variables.


Based on the Chi-Square test, Bar plot visualization and Heatmap, **there is a statistically significant relationship between Pclass and Survival**, with higher survival rates associated with higher classes.

Scaling and normalization are crucial preprocessing steps in many machine learning projects because they ensure that the features are on a similar scale and improve the performance and convergence of various models. It helps:

- Improves Visualization and Interpretability

However, If we are unsure about the case, the best Practice is scaling the features is a safe and recommended approach for consistent results.

Logistic Regression is a **linear classification algorithm** that predicts the probability of a binary outcome (e.g., `Survived` or `Not Survived`) based on one or more input features. ### 5.4.3 **Perceptron**

### Description of the Model

The Perceptron is a **linear classification algorithm** inspired by biological neurons. It is the simplest type of neural network and works by finding a hyperplane that separates data into two classes. The Perceptron is useful for binary classification problems but assumes that the data is **linearly separable**. The model is **simple, interpretable, and computationally efficient**.

It assumes:

- **Linear Decision Boundary**: Works well when the data can be separated by a straight line (or hyperplane in higher dimensions).
- **Threshold-Based Predictions**: Outputs binary predictions directly (e.g., 0 or 1) rather than probabilities.
- **Simple and Fast**: Very efficient to train and interpret but less powerful than modern algorithms.
- **Scaling**: Scaling continuous features is important to ensure faster convergence and better performance.

### How It Works:

- The Perceptron models the relationship between input features (e.g., \(x_1, x_2, ..., x_n\)) and the target variable (e.g., class labels 0 or 1).
- It calculates a **weighted sum** of the input features and passes it through a threshold activation function:
  - If the weighted sum is **greater than 0**, the output is predicted as **1 (positive class)**.
  - Otherwise, it predicts the output as **0 (negative class)**.
- During training, the Perceptron adjusts its weights iteratively using the **Perceptron Learning Rule**:
  - For every misclassified example, the weights are updated.
- The algorithm **repeats until all training examples are classified correctly** (if the data is linearly separable) or a maximum number of iterations is reached.

### 5.4.4 **Stochastic Gradient Descent (SGD)**

### Description of the Model
**Stochastic Gradient Descent (SGD)** is a versatile optimization algorithm widely used in training machine learning models, including linear regression, logistic regression, and deep learning. **It updates model parameters iteratively to minimize the loss function**. Unlike traditional Gradient Descent, which processes the entire dataset at once, SGD **updates parameters for each individual data point**, enhancing **efficiency and scalability** for large datasets.

Key assumptions include:
- **Incremental Learning:** Processes data incrementally, making it suitable for large datasets and reducing memory usage.
- **Convex Loss Function:** Ensures effective convergence for convex functions, achieving global minima.
- **Learning Rate Sensitivity:** The learning rate must balance convergence speed and stability to prevent divergence.
- **Noise Tolerance:** The stochastic nature of updates introduces randomness, which aids in escaping local minima in non-convex functions.
- **Regularization Support:** Incorporates techniques to mitigate overfitting, further improving generalization.

Despite its advantages, SGD can exhibit instability near convergence due to oscillations, requiring careful learning rate tuning. Its reliance on single-sample gradients introduces noise that may slow convergence, and hyperparameter selection remains critical for optimal performance.

### How It Works:
1. **Initialization:**
   - Start with random or zero values for model parameters (weights, biases).
2. **Iterative Updates:**
   - For each training sample (or batch in Mini-Batch SGD):
     - Calculate the gradient of the loss function relative to the parameters.
     - Update parameters using the formula that using Old model parameters (weight, biases) deduct by the multiplication of learning rate (step size) and Gradient of the loss function*.
3. **Repeat Until Convergence:**
   - Iterate through all training samples over multiple itreation (epochs) until the loss stabilizes or predefined stopping criteria (e.g., maximum epochs, minimum loss) are met.

### Applied to Titanic Dataset:
- The **Titanic dataset** includes features such as **Pclass**, **Sex**, **Age**, **Fare**, and **Embarked**.
- SGD can be used to train models predicting passenger survival (1) or non-survival (0).
- Features like **Sex (binary)** and numerical attributes such as **Age** and **Fare** contribute to effective loss minimization during training.
- Its iterative update mechanism enables efficient handling of large-scale datasets, especially when combined with feature scaling and preprocessing.

*-The loss function is a mathematical formula that measures how well a model performs by comparing its predicted output to the actual output. A higher loss value indicates greater errors in the model's predictions.*

*-The gradient is a vector that indicates the direction and magnitude of parameter adjustments required to minimize the loss as quickly as possible. It is calculated by taking the derivative of the loss function with respect to the model's parameters.*

*-The gradient of the loss function provides guidance on how to adjust the model's parameters (e.g., weights or biases) to reduce the loss function value in each iteration of the learning process.*

### 5.4.5 **Decision Tree**

### Description of the Model
Decision Trees are a versatile machine learning model **used for both classification and regression tasks**. They work by recursively splitting the dataset based on feature values to create a **tree-like structure**, where each node represents a decision rule, and leaf nodes represent outcomes (e.g., a class label or prediction). 

Unlike linear models, **Decision Trees can capture non-linear relationships** and effectively **handle both categorical and numerical features**. Their interpretable structure makes them a popular choice for understanding the decision-making process of a model.

Key assumptions include:
- **Non-Linear Splits**: Decision Trees do not assume a linear relationship between features and the target variable, making them suitable for complex datasets.  
- **Feature Importance**: Features are split based on their ability to reduce impurity (**measured using Gini Impurity or Entropy, with the goal of minimizing these criteria to values close to 0**) at each node.

  - When using **Gini Impurity** as the criterion for evaluating the quality of splits in a tree-based model, it measures the **degree of impurity or misclassification** in the data. The goal is to achieve the **lowest impurity** or the **highest purity** in its nodes. Lower impurity indicates that the model can effectively separate the classes within the nodes, resulting in more homogeneous groups. Conversely, higher impurity indicates that the node contains a mix of classes, leading to poorly defined separations.

  - When using **Entropy** as the criterion, it measures the **uncertainty or disorder** in the data. The goal is to minimize entropy to **0**, which represents **maximum certainty** and indicates that the information in the nodes is well-ordered and predominantly from one class. In contrast, entropy closer to **1** represents **maximum uncertainty**, meaning the data in the nodes is highly disordered and contains an equal mix of different classes, making separation difficult.

- **Iterative Splitting**: The model iteratively splits data until a stopping condition is met (e.g., maximum depth or minimum samples per leaf).  
- **Overfitting Risk**: Decision Trees **can overfit the training data if the tree is too deep**, requiring regularization techniques like **pruning or setting depth limits**.

Despite their strengths, Decision Trees can suffer from instability when trained on small datasets or noisy data, and are sensitive to small changes in the dataset. They are often combined with ensemble methods (e.g., Random Forests) to improve robustness and accuracy.

### How It Works

1. **Initialization**:
   - Start with the entire dataset and all available features.

2. **Splitting Data**:
   - For each feature, evaluate potential splits by measuring how well they reduce impurity (e.g., Gini Impurity or Entropy).
   - Choose the best split that maximizes information gain and create two child nodes.

3. **Recursive Splitting**:
   - Repeat the splitting process for each child node, using only the data assigned to that node.
   - Continue splitting until stopping criteria are met (e.g., maximum tree depth, minimum samples per leaf, or no further impurity reduction).

4. **Stopping Condition**:
   - Once the stopping criteria are met, assign the majority class (classification) or mean value (regression) of the data in each leaf node as the final prediction.

### Applied to Titanic Dataset  

- **Dataset Features**:  
  The Titanic dataset includes features such as **Pclass, Sex, Age, Fare, and Embarked**. Decision Trees can handle these features directly, as they process both categorical and numerical data without the need for one-hot encoding or normalization.

- **Decision Rules**:  
  Splits are created based on questions such as "Is the passenger male?" or "Is the passenger in 3rd Class?" to group passengers with similar survival probabilities.

- **Outcome Examples**:  
  Leaf nodes predict survival probabilities or binary outcomes:
  - **Female, 1st/2nd Class**: High survival probability (e.g., 90%).  
  - **Male, 3rd Class**: Low survival probability (e.g., 10%).  

- **Efficiency and Scalability**:  
  The recursive splitting mechanism allows efficient handling of datasets, particularly when combined with constraints like maximum depth to prevent overfitting.

*Gini Impurity and Entropy are metrics used in decision trees to measure how "pure" a dataset is at a given split:*

*- Gini Impurity: Measures the likelihood of incorrectly classifying a randomly chosen data point if it were labeled based on the class distribution in a split.*

*- Entropy: Measures the uncertainty or disorder in the data at a given split. Suitable for more complex data*

| **Criterion**           | **Gini Impurity**                                                                 | **Entropy**                                                        |
|--------------------------|----------------------------------------------------------------------------------|--------------------------------------------------------------------|
| **Measurement**          | Measures impurity or misclassification.                                          | Measures uncertainty or disorder.                                  |
| **Goal**                 | Minimize impurity to achieve node purity.                                        | Minimize uncertainty to achieve high order.                       |
| **Computation**          | Simpler and faster (does not require logarithms).                                | More complex (requires logarithms).                               |
| **Range**                | 0 to 0.5 (binary classification).                                                | 0 to 1.                                                           |
| **Interpretability**     | Easier to understand and interpret.                                              | May be more complex to interpret.                                 |
| **Performance**          | Provides similar performance in practice for general use cases.                 | Slightly better suited for more complex statistical analyses.      |
| **Suitability**          | Works well for general cases.                                                    | Suitable for detailed data analysis with more complexity.          |
| **Computation Speed**    | Faster due to simpler calculation.                                               | Slower due to logarithmic calculations.                           |

### 5.4.6 **Random Forest**

### Description of the Model
A **Random Forest Model** is a powerful machine learning algorithm that **combines multiple decision trees** to make predictions. Each tree makes an independent decision based on the data, and the final outcome is determined by the majority vote of all the trees. This ensemble approach increases accuracy, reduces overfitting, and makes the model more robust.

Random Forest can be thought of as a group of decision-making experts (trees), each analyzing the passenger data (e.g., gender, ticket class, siblings/spouses aboard) from different perspectives. By combining the decisions of all trees, the model improves accuracy and avoids over-reliance on any one feature. 

It is an **ensemble learning method**, which reduces the likelihood of overfitting and helps to handle noisy or complex data. The advantages of the Random Forest Model include:

1. **Accuracy**: Combining predictions from multiple trees reduces errors compared to a single tree.
2. **Diversity**: Randomly selecting subsets of data and features helps avoid overfitting.
3. **Robustness**: The model performs well even with noisy or missing data, as no single tree dominates the decision-making process.

### Step 1: Data Preprocessing
Before building the model, the data needs to be preprocessed. This involves handling missing values, encoding categorical variables (such as gender), and scaling numerical features if necessary.

### Step 2: Building a Decision Tree

A **decision tree** is a model that makes decisions by splitting data into smaller, more homogeneous groups. The tree identifies the best splits for each feature to maximize the difference between survivors and non-survivors in the Titanic dataset.

#### **1. First Split: Gender**
The tree starts by evaluating the `Sex` feature:
- If the passenger is **female**, the tree moves down one branch.
- If the passenger is **male**, the tree moves down another branch.

This is an important split because **gender** had a significant influence on survival, with women being prioritized during rescue operations.

#### **2. Second Split: Siblings/Spouses (`SibSp`)**
For female passengers, the tree then examines the number of siblings or spouses aboard (`SibSp`):
- If **`SibSp <= 3`**, the tree moves to the next split.
- If **`SibSp > 3`**, the tree predicts **“Did Not Survive”** (`Survived = 0`).

The threshold **`SibSp <= 3`** was found to be a good separator for predicting survival based on family size, where smaller families had higher survival rates.

#### **3. Third Split: Parents/Children (`Parch`)**
For female passengers with **`SibSp <= 3`**, the tree evaluates the `Parch` feature (number of parents or children aboard):
- If **`Parch < 4`**, the tree predicts **“Survived”** (`Survived = 1`).
- If **`Parch >= 4`**, the tree predicts **“Did Not Survive”** (`Survived = 0`).

The **`Parch < 4`** threshold suggests that passengers with smaller families had better survival chances.

#### **How Does the Tree Choose These Thresholds?**
The decision tree algorithm automatically selects the thresholds by:
1. Testing different possible splits for each feature.
2. Using criteria like **Gini Impurity** or **Entropy** to evaluate how well each split separates survivors from non-survivors.
3. Choosing the split that results in the most homogeneous groups (with most passengers either surviving or not).

Each tree in the Random Forest learns different rules based on its data subset, leading to unique trees with different splits.

*Note: This describes the decision process of one tree. A Random Forest consists of multiple such trees, each making its own decisions.* 

### Step 3: Voting Mechanism

Once the Random Forest has built multiple decision trees, it aggregates their predictions to determine the final outcome.

#### **How Voting Works:**
1. Each tree makes an independent prediction for the new passenger.
2. The Random Forest counts the predictions from all trees using **majority voting**.
3. The final prediction is the **outcome with the most votes**.

#### **Example of Voting**
For a new passenger (`Sex = female`, `Pclass = 3`, `SibSp = 1`, `Parch = 0`), the trees in the Random Forest might predict as follows:

- **Tree #1**: Predicts **Survived** (`1`).
- **Tree #2**: Predicts **Did Not Survive** (`0`).
- **Tree #3**: Predicts **Survived** (`1`).

The votes are tallied:
- **2 votes for “Survived”**.
- **1 vote for “Did Not Survive”**.

**Final Prediction**: The passenger is predicted to have **survived**, as it received the majority vote.

*Thus, the final prediction for each passenger is determined by the majority vote of all the trees in the Random Forest.*


### Exploring `n_estimators` and `max_depth`

When performing the Random Forest model, different accuracy scores are obtained by varying the hyperparameters **Number of trees in the forest (`n_estimators`)** and **Maximum depth of trees (`max_depth`)**.

#### **Effect of `n_estimators` (Number of Trees)**
- Increasing the number of trees improves model accuracy, especially in the early stages, by helping the model better capture patterns in the data and reducing overfitting.
- However, beyond a certain point (e.g., 100-200 trees), accuracy plateaus as additional trees no longer add significant new information.
- **Benefits**:
  - Reduces variance in predictions, resulting in more stable and reliable outputs.
  - Helps generalize the model to unseen data.
- **Limitations**:
  - Beyond the plateau, adding more trees increases computational cost without significant accuracy gains or it might decrease accuracy score.

#### **Effect of `max_depth` (Maximum Tree Depth)**
The maximum depth controls the trade-off between **node purity** and **model complexity**:
- **Higher Depth**:
  - Captures more detailed patterns, especially with `max_depth=None`, achieving maximum node purity.
  - However, **deeper trees risk overfitting by learning unnecessary noise**, reducing performance on new data and increasing computational demands.
- **Lower Depth**:
  - **Limits the tree's ability to split excessively**, but reducing complexity and overfitting.
  - Promotes **better generalization**, as the model focuses on meaningful patterns rather than noise.
  - **Risk**: Too shallow trees may underfit, failing to capture important details in the data.

#### **Key Observations**
- **Overfitting Prevention**: Limiting `max_depth` reduces the model's focus on minor details, improving generalization to unseen data.
- **Underfitting Risk**: Too small `max_depth` may miss key data patterns, lowering model performance.

### 5.4.7 K-Nearest Neighbors (KNN)

### Description of the Model

K-Nearest Neighbors (KNN) is a simple yet powerful **non-parametric supervised machine learning algorithm** used for both **classification** and **regression** tasks. The model makes predictions based on the **similarity of a data point to its closest neighbors** in the feature space. KNN does not make any assumptions about the underlying data distribution, making it versatile for many applications.

Unlike some complex models, KNN relies on **distance metrics** to measure the similarity between points and determine the most likely class or value of an unknown point based on its neighbors. It is particularly **effective for small datasets with normalized features.**

Key assumptions include:

1. **Similarity Matters**: Data points that are closer in the feature space are more likely to belong to the same class or have similar target values.
2. **Feature Normalization**: Features must be scaled (e.g., using Min-Max scaling or Standardization) to ensure all features contribute equally to distance calculations.
3. **Instance-based Learning**: KNN does not involve explicit training. Predictions are made at runtime by comparing the query point to the stored training examples.
4. **Choosing 'K'**: The number of neighbors ('K') plays a critical role, A **larger value of K means the model considers a broader range (greater number of nearby neighbors or perspective)** by including more neighbors in its decision-making process. When you want to predict the class or value of a new data point, the model looks at the K 'neighbors' (the closest data points) in the nearby area (called the feature space) and uses information from those neighbors to make a prediction.

### How Does the Choice of 'K' Affect the Model?

| **K Value**       | **Description**                                                                                     | **Advantages**                                   | **Potential Issues**                              |
|--------------------|-----------------------------------------------------------------------------------------------------|-------------------------------------------------|--------------------------------------------------|
| **Small K** (e.g., K=1 or K=3) | Focuses on the closest neighbors, using only a few nearby data points.                                      | - High accuracy in densely packed regions.      | - Overfitting: Sensitive to noise or outliers.   |
| **Large K** (e.g., K=10 or K=20) | Considers a broader neighborhood, looking at many neighbors in the dataset.                              | - Reduces sensitivity to noise.                | - Underfitting: Misses fine-grained patterns.    |

### Additional Note:
- Small values of 'K' can lead to **overfitting**, while large values can smooth out predictions, potentially missing important details in the data.


Despite its simplicity, KNN is computationally intensive, especially for large datasets, as it requires calculating distances for all training points during predictions.

### How It Works

1. **Initialization**:
   - Choose the number of neighbors (**K**) to consider for predictions.
   - Prepare the dataset by normalizing feature values to ensure equal contribution to distance calculations.

2. **Distance Calculation**:
   - For a given query point, calculate the distance between the query and all training points in the dataset. Common distance metrics include:
     - **Euclidean Distance**: Straight-line distance, Sensitive to large values, suitable for Low-dimensional space
     - **Manhattan Distance**: Along axes/grid, Less sensitive, High-dimensional/sparse
     - **Minkowski Distance**: Customizable, Generalized form of both Euclidean and Manhattan distances.

3. **Neighbor Selection**:
   - Identify the **K closest neighbors** to the query point based on the chosen distance metric.

4. **Prediction**:
   - **For Classification**:
     - Count the classes of the K neighbors.
     - Assign the majority class as the predicted label.
   - **For Regression**:
     - Calculate the average or weighted average of the target values of the K neighbors and use it as the predicted value.

### Applied to Titanic Dataset

#### Dataset Features:
The Titanic dataset includes features such as **Pclass, Sex, Age, Fare**, and **Embarked**. These features are normalized before applying the KNN algorithm to ensure accurate distance calculations.

#### Decision Rules:
- KNN identifies neighbors based on similarities in feature values. For example, passengers with similar **Pclass** and **Fare** might be closer neighbors in the feature space.

#### Outcome Examples:
- **Classification**: Predict whether a passenger survived (binary classification: survived vs. not survived) by analyzing the survival outcomes of their closest neighbors.
- **Regression**: Predict the fare a passenger might pay based on the fares of their closest neighbors.

### Advantages of KNN
- Simple to understand and implement.
- Flexible for both classification and regression tasks.
- No assumptions about data distribution.

### Limitations of KNN
- Computationally expensive for large datasets.
- Sensitive to the choice of distance metric and the value of 'K'.
- Performance depends on well-normalized features and a representative training set.

### 5.4.8 **Gradient Boosting Classifier**

### **Description of the Model**
Gradient Boosting Classifier is a powerful machine learning technique that builds an **ensemble of weak prediction models, typically decision trees, to create a stronger predictive model**. It is an iterative process where models are trained sequentially, and each new model focuses on correcting the errors of the previous ones. By combining the predictions of multiple models, Gradient Boosting **minimizes the overall prediction error and achieves high accuracy**.

Key features of Gradient Boosting include:
- **Sequential Learning**: Each model is trained to correct the residual errors of the prior models, ensuring continuous improvement.
- **Additive Model**: Predictions are combined in a weighted manner to achieve better performance than any individual model.
- **Loss Function Optimization**: The algorithm minimizes a user-specified loss function (e.g., log loss for classification or mean squared error for regression).
- **Learning Rate**: A hyperparameter that controls the contribution of each model to the final prediction, preventing overfitting.
- **Regularization**: Techniques like tree depth constraints, minimum samples per leaf, and shrinkage are used to reduce overfitting and improve generalization.

While Gradient Boosting provides excellent performance, it may be computationally expensive and sensitive to hyperparameter tuning.

### **How It Works**
1. **Initialization**:
   - Start with a weak model (e.g., a simple decision tree) that predicts a constant value, such as the mean of the target variable.

2. **Residual Computation**:
   - Calculate the residuals (errors) between the true labels and the predictions made by the current model.

3. **Train a Weak Model on Residuals**:
   - Fit a new weak model (e.g., a decision tree) to the residuals. This weak model attempts to minimize the remaining error.

4. **Update Predictions**:
   - Combine the predictions of the weak model with the predictions of the ensemble using a weighted learning rate.

5. **Repeat**:
   - Iteratively add weak models to the ensemble until a stopping criterion is met (e.g., a maximum number of models or no significant improvement in performance).

6. **Final Prediction**:
   - Combine the predictions of all weak models in the ensemble to generate the final prediction.

### **Applied to Titanic Dataset**
The **Titanic Dataset** includes features such as `Pclass`, `Sex`, `Age`, `Fare`, and `Embarked`, which are used to predict survival (`1` for survival, `0` for non-survival). Gradient Boosting Classifier can be applied to this dataset as follows:

1. **Feature Engineering**:
   - Process categorical features (e.g., `Sex`, `Embarked`) and scale numerical features (e.g., `Age`, `Fare`) to ensure effective training.

2. **Model Training**:
   - Use Gradient Boosting to train a classifier on the Titanic dataset, minimizing a classification loss function (e.g., log loss).

3. **Advantages**:
   - **Handles Mixed Data Types**: Efficiently handles both categorical and numerical features.
   - **Feature Importance**: Provides insights into the importance of features in predicting survival.
   - **High Accuracy**: Performs well on imbalanced datasets with proper hyperparameter tuning.

4. **Performance Considerations**:
   - Carefully tune hyperparameters such as the number of estimators (trees), learning rate, and tree depth to prevent overfitting or underfitting.
   - Use techniques like cross-validation to validate model performance.


Gradient Boosting Classifier is particularly well-suited for the Titanic dataset due to its ability to capture complex relationships and interactions between features, resulting in high predictive accuracy when properly tuned.

### 5.4.9 **Support Vector Machines**

### Description of the Model
Support Vector Machines (SVM) is a powerful supervised learning algorithm used for classification and regression tasks. SVM aims to find the optimal hyperplane that separates data points of different classes with the maximum margin. By maximizing the margin, SVM enhances the generalization of the model, making it robust to unseen data. For non-linear data, SVM uses kernel functions to map data to higher dimensions, enabling effective classification in complex scenarios.

**Key assumptions include:**
- **Separability of Classes**: Data can be linearly or non-linearly separable in the feature space.
- **Margin Maximization**: The algorithm focuses on maximizing the margin between classes to improve decision boundary stability (Hyperplane).
- **Use of Support Vectors**: Only the closest data points (support vectors) to the decision boundary affect its placement.
- **Kernel Trick**: Kernels like linear, polynomial, and radial basis function (RBF) transform non-linear data into higher dimensions for separability.

Despite its advantages, SVM can be computationally intensive for very large datasets and sensitive to the choice of hyperparameters such as the kernel type, regularization parameter (C), and gamma (for RBF). Proper tuning and scaling of input features are essential for optimal performance.

### How It Works:
1. **Input Feature Space and Data Preparation**:  
   - Input data is preprocessed and scaled to improve algorithm performance and avoid dominance of features with larger numerical ranges.

2. **Hyperplane Optimization and finding the margin***:  
   - SVM calculates the hyperplane* that maximizes the margin between two classes in a binary classification problem.
   - **Margin Maximization** means increasing the distance (margin) between the hyperplane or **maximizing the distance between the hyperplane (decision boundary) and the closest data points from each class (support vectors)** and the closest data points from each class as much as possible. **The Best Hyperplane is the one that maximizes this margin**, making the model more flexible and better at handling new data.

3. **Kernel Transformation (if needed)**:  
   - For non-linearly separable data, a kernel function is applied to map input features to a higher-dimensional space.

4. **Regularization and Optimization**:  
   - The regularization parameter (C) balances the trade-off between maximizing the margin and minimizing classification errors.

5. **Classification or Regression**:  
   - After training, Support Vector* Machines predicts the class or value of new data points based on their position relative to the hyperplane.

### Applied to Titanic Dataset
- The **Titanic dataset** includes features such as **Pclass, Sex, Age, Fare, and Embarked**.
- SVM can be used to predict passenger survival (1) or non-survival (0) with high accuracy.
- Features like **Sex (binary)** and numerical attributes such as **Age** and **Fare** can be normalized and used to train the SVM model effectively.
- The kernel trick (e.g., RBF or polynomial kernels) can improve classification for non-linear relationships between features and outcomes.
- SVM’s capability to handle both linear and non-linear data makes it suitable for complex datasets with overlapping or non-linear patterns.

### Data Preparation
- Remove redundant or less relevant features, even if they seem meaningful or highly correlated. They might add unnecessary complexity without significantly improving model performance.
- Figuring out which features are redundant depends heavily on individual judgment.
- A good practice is to establish a baseline model without "feature crosses" first, then experiment by adding them one by one with the primary goal of improving the score.
- Some features are better kept as continuous, such as `Fare`. Keeping it continuous:
  - **Works Well Across Models**: Continuous values work well for algorithms like Logistic Regression and Gradient Boosting. Tree-based models like Random Forests can handle both continuous and categorical data without requiring manual binning
  - **Saves Effort**: Avoids the additional step of defining and tuning bins, which can be arbitrary and might misalign with the dataset's structure.
- Outlier management: Instead of binning, handle outliers through scaling or clipping (e.g., log transformation or capping high values).
- Scaling requirements:
  - **Crucial for**: Logistic Regression, KNN, SGD, Perceptron.
  - **Not required for**: Decision Tree, Random Forest, Gaussian Naive Bayes.
  - **Optional for Linear Regression**: Helps in specific cases but is not mandatory.

### Model Training
- What you think might work best isn’t always the case—trial and error, like with Random Forests, taught me valuable lessons about making and testing assumptions.
- Some models, like Logistic Regression, prefer simplicity over granularity. For example:
  - Binning `Age` into categories simplifies splits and reduces noise from outliers.
  - Simplifying data, such as Binning Fare can simplify the data but risks losing valuable information. It should be tested carefully as it may affect model performance unpredictably. However, always start with a baseline model before introducing such changes.
- Simplification can sometimes result in a loss of valuable information. Running experiments with and without simplified features can provide insights into their actual contribution to the model's performance.
- Continuous features like `Fare` are generally preferred, especially for models that can directly handle them.
- Ensure proper scaling for models that require it (e.g., Logistic Regression, KNN, SGD, Perceptron). Use `StandardScaler` or `MinMaxScaler` before fitting these models.

This methods will show how to use **Cross-Validation (suitable for Small Datasets)** becuase when the dataset is small, **splitting it into train and test sets may leave insufficient data** for training. In such cases, cross-validation is a better approach. Cross-validation provides a reliable measure of performance by testing the model on multiple folds of data, ensuring every data point is used for both training and testing.

### Steps to Evaluate and Build the Model
1. **Train and Predict on the Training Dataset**:
   - Initially, train the model on the entire dataset and evaluate its accuracy on the same training data. While this accuracy will be high, it reflects only how well the model memorizes the training data.

2. **Compare Accuracy from Training Dataset**:
   - Compare the model's accuracy when predicting on the training set among multiple models and select only the highest accuracy score

3. **Perform K-Fold Cross-Validation**:
   - Use K-fold cross-validation to reflect the true accuracy of the model. This process splits the training data into multiple folds, testing the model’s ability to generalize across unseen subsets of the data. 
   - The mean accuracy score from cross-validation provides a more reliable metric for generalization and we will use this figures compared to the accuracy score from training dataset. The figurs will show **a significant drop in accuracy** due to overfitting.


4. **Deploy to Test Dataset**:
   - After selecting the best model based on cross-validation results, evaluate it on the reserved test dataset to simulate real-world performance and compute key performance indicators (KPIs).
